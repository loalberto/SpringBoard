**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

```python
# This will transpose the original dataframe
transposed_properties = properties.T
# The method reset_index will reset the data to 0 to size of dataframe -1
transposed_properties.reset_index(inplace=True)
# The columns method will set the values to the first row of the dataframe.
transposed_properties.columns = transposed_properties.iloc[0]
transposed_properties = transposed_properties.drop(0)
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
# Renaming the columns to their appropriate names
transposed_properties = transposed_properties.rename(columns={'Unnamed: 0': 'London_Boroughs', pd.NaT: 'ID'})
transposed_properties.head()
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
# The idea is to make all the columns of dates apart of the value data. I will take them and convert them into rows
# of data since they are there to describe our data.

# I will retrieve the dates after 'ID'
lis_columns = transposed_properties.columns[2:]
# Melt the dataframe, keeping 'London_Boroughs' and 'ID', into a table wiht only 4 columns of data.
melted_properties = pd.melt(transposed_properties, id_vars=['London_Boroughs', 'ID'], value_vars=lis_columns)
melted_properties.head()
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
# I have renamed the column names to be 'Month' and "Average_Prices" as the guidline says below.
melted_properties = melted_properties.rename(columns={0: 'Month', 'value': 'Average_Prices'})
melted_properties.head()
```

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit.

```python
# These lists will contain the indices of where the values with null values exist 
null_value_list_boroughs = []
# Series were made to study the london_column
london_boroughs = melted_properties['London_Boroughs']

# Iterate through each value in each row
for i in range(len(london_boroughs)):
    # If it contains the value 'Unnamed' then append the index to its appropiate list
    if 'Unnamed' in london_boroughs[i]:
        null_value_list_boroughs.append(i)

# Finally, dropping the rows with null values 
non_null_dataframe = melted_properties.drop(null_value_list_boroughs)
# This reset will bring the index back to a sequential order
non_null_dataframe = non_null_dataframe.reset_index(drop=True)
# The values appear to be matching
non_null_dataframe.info()
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
def create_price_ratio(df):
    # Dictionary to hold all of the ratio data
    t_1998 = []
    t_2018 = []
    r_total = []
    borough_list = []
    # This set will get the unique borough values in the dataset
    unique_boroughs = set(df['London_Boroughs'])
    # The for loop will iterate through each borough to extract each year
    for borough in unique_boroughs:
        borough_frame = df[df['London_Boroughs'] == borough]
        # It will then filter out the years of 1998 and 2018
        price_of_1998 = borough_frame[borough_frame['Year'] == 1998]
        price_of_2018 = borough_frame[borough_frame['Year'] == 2018]
        # Using a map & lambda, I was able to obtain all the average prices a given year
        total_1998 = list(map(lambda v: v, price_of_1998['Average_Prices']))
        total_2018 = list(map(lambda v: v, price_of_2018['Average_Prices']))
        
        # Then performed the ratio conversion
        total_1998 = sum(total_1998) / len(total_1998)
        total_2018 = sum(total_2018) / len(total_2018)
        
        ratio = total_2018 / total_1998
        # Appends each borough and ratio value
        t_1998.append(total_1998)
        t_2018.append(total_2018)
        r_total.append(ratio)
        borough_list.append(borough)
    # Creates it into a temporary series
    df_dict = {'Borough': borough_list, '1998_Average': t_1998, '2018_Average': t_2018, 'Ratio': r_total}
    df = pd.DataFrame(df_dict)
    return df
# The final dataframe gets created
final_df = create_price_ratio(df)
final_df
```
